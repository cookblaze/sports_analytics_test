```python
import pandas as pd
import numpy as np
import pybaseball
from pybaseball import statcast
from sklearn.linear_model import LogisticRegression
```

```python
data = statcast(start_dt='2019-04-26', end_dt='2019-04-26')
```

```python
data.head(50)
```

```python
data.columns.tolist()
```

```python
data[["estimated_ba_using_speedangle","events","pitch_type","batter","fielder_2.1","fielder_3","fielder_4","fielder_5","fielder_6","fielder_7","fielder_8","fielder_9"]].dropna()
```

```python
test = data[["estimated_ba_using_speedangle","events","pitch_type","batter","fielder_2.1","fielder_3","fielder_4","fielder_5","fielder_6","fielder_7","fielder_8","fielder_9","launch_speed"]].dropna()
```

```python
#define what is a hit and encode it as a 1 or 0
hits = ['home_run', 'single', 'double', 'triple']
test["is_hit"] = test["events"].isin(hits).astype(int)
```

```python
pd.get_dummies(test)
```

```python
clf = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial').fit(pd.get_dummies(test.drop(["is_hit","events"],axis=1)), test["is_hit"])
```

```python
clf.score(pd.get_dummies(test.drop(["is_hit","events"],axis=1)), test["is_hit"])
```
